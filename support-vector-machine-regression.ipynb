# Support Vector Machine - Regression

We have talked about classification quite a bit in this class, but lets look at how we can leverage support vector machines for a regression problem. 

This is a simple example how how you might use a Support Vector Machine Regression. 

From the previous notebook we learned how we want to keep points outside the margin to provide a classification, for regression we can flip that on its head. For the regression model the idea is to keep every point within the margin, that way the "dividing line" can be used as the regression line. 

Notebook Citations:
- https://scikit-learn.org/1.5/modules/generated/sklearn.svm.SVR.html
- https://www.geeksforgeeks.org/support-vector-regression-svr-using-linear-and-non-linear-kernels-in-scikit-learn/

```python
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.svm import SVR 
from sklearn.model_selection import train_test_split 
%matplotlib ipympl

X = np.sort(np.random.uniform(-1, 1, 200))
y = np.sin(4 * X) + np.random.randint(-100, 100, 200)/200

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.25, random_state=10)
plt.figure()
plt.scatter(X, y)
plt.plot(X_train, y_train, 'rx')
plt.show()
```

```python
kernel_types = ['linear', 'poly', 'rbf']

fig, ax = plt.subplots(2, 2)
for index, k_type in enumerate(kernel_types):
    model = SVR(kernel=kernel_types[index])
    model.fit(X_train.reshape(-1, 1), y_train.reshape(-1, 1))
    y_hat = model.predict(X_test.reshape(-1, 1))
    row = int(index/2)
    column = index%2
    ax[row, column].scatter(X_test, y_test)
    ax[row, column].plot(X_test, y_hat, 'xr')
    ax[row, column].set_ylabel(k_type)
    ax[row, column].set_xlabel(round(model.score(X_test.reshape(-1, 1), y_test.reshape(-1, 1)), 3))
    
plt.show()

```
